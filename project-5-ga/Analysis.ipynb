```python
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(20,20))
ax = plt.subplot()
plt.xlabel("Fitness Evaluation for Generation"); plt.ylabel("Best Fitness for Generation")
for i in range(1, 31):
    df = pd.read_csv("HillClimber" + str(i), delimiter = ",", skiprows = [0])
    x = df["FitnessEval"]
    y = df["BestFitness"]
    ax.plot(x, y)
```
